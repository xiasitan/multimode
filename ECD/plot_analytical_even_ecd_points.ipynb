```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# define parameters
alpha = 2
t1 = 1
t2 = 1
chi_g = 0.1
chi_e = 0.3

# define points
g = alpha*(1-2*np.cos(chi_g*t2)+np.cos(chi_e*t1-chi_g*t2)) + 1j*alpha*(2*np.sin(chi_g*t2)-np.sin(chi_g*t2-chi_e*t1))
e = alpha*(1-2*np.cos(chi_e*t2)+np.cos(chi_g*t1-chi_e*t2)) + 1j*alpha*(-2*np.sin(chi_e*t2)-np.sin(chi_g*t1-chi_e*t2))

# calculate the midpoint of g and e
midpoint = (g+e)/2

# translate the points by the midpoint
g_translated = g - midpoint
e_translated = e - midpoint

# calculate the phase with the y axis of both poitns
phase_g = np.angle(g_translated)
phase_e = np.angle(e_translated)

# rotate both points by the phase
g_rotated = g_translated*np.exp(-1j*(phase_e+np.pi/2))
e_rotated = e_translated*np.exp(-1j*(phase_e+np.pi/2))

# plot the points and a corrinate system
plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.plot(np.real(g), np.imag(g), 'go', label='g')
plt.plot(np.real(e), np.imag(e), 'bo', label='e')
plt.plot(np.real(g_translated), np.imag(g_translated), 'g*', label='g_translated')
plt.plot(np.real(e_translated), np.imag(e_translated), 'b*', label='e_translated')

plt.plot(np.real(e_rotated), np.imag(e_rotated), 'g+', label='e rotated')
plt.plot(np.real(g_rotated), np.imag(g_rotated), 'b+', label='g rotated')

plt.show()
```
