## 2.2 Plot Wigner

### Review: Wigner quasi-probability distribution

$ W (\beta) = \frac{2}{\pi} \text{Tr} \left( D^\dagger(\beta) \rho D(\beta) \hat{P} \right) $ with $ \hat{P} = e^{~i \pi \hat{n}} $

`wigner(psi, xs, ys, g=2)`

psi: state to visualize

xs: array of x-coordinates to evaluate the Wigner function at

ys: array of y-coordinates to evaluate the Wigner function at

g: a scaling factor (use g=2.0 , default is g=sqrt(2))

Here, instead of calculating the Wigner function using `wigner()`, we visualize it using the helper function `qutip.visualization.plot_wigner`

### 2.2.1 Two-Mode Wigner Tomography

1. NOON state with $N=5$ shared between Alice and Bob: $~~~~|\text{NOON} \rangle_{AB} = \frac{|5,0\rangle_{AB} ~+~ |0,5 \rangle_{AB}}{\sqrt{2}}$ 

2. $\rho_A = \text{tr}_B \left( |\text{NOON} \rangle_{AB} \langle \text{NOON} | \right) = \frac{1}{2}  |0\rangle \langle 0|  + \frac{1}{2}  |5\rangle \langle 5| = \rho_B $

### Wrong Wigner: plugging the whole state into the $wigner$ function

```python
dim = 6
noon = ( tensor( fock(dim,5), fock(dim,0) ) + 
        tensor( fock(dim,0), fock(dim,5) ) ).unit()

# Initalize plot variables
xvec = linspace(-4,4,200)
yvec = linspace(-4,4,200)

wi = wigner(noon, xvec, yvec, g = 2.0) # noon is a 2-mode state, wigner only forks for single modes

## ploting setting
fig, axes = plt.subplots(1, 1, figsize=(10, 4)) # Create a 1-plot figure
# Contour plots using the results of the Wigner function
cont0 = axes.contourf(xvec, yvec, wi, 200, cmap = 'seismic', norm = plt.Normalize(vmin = -2/pi, vmax = 2/pi))
axes.set_aspect(1) # Make plots squared
fig.colorbar(cont0, ax = axes) # Add colorbar
lbl0 = axes.set_title("Noon") # set the title for each plot
```

### Alice Wigner: need to trace out Bob

```python
dim = 6
noon = ( tensor( fock(dim,5), fock(dim,0) ) + tensor( fock(dim,0), fock(dim,5) ) ).unit()

# Initalize plot variables
xvec = linspace(-4,4,200)
yvec = linspace(-4,4,200)

wi = wigner(noon.ptrace(0), xvec, yvec, g = 2.0) # Calculate the Wigner function for the states

fig, axes = plt.subplots(1, 1, figsize=(10, 4)) # Create a 1-plot figure
# Contour plots using the results of the Wigner function
cont0 = axes.contourf(xvec, yvec, wi, 200, cmap = 'seismic', norm = plt.Normalize(vmin = -2/pi, vmax = 2/pi))
axes.set_aspect(1) # Make plots squared
fig.colorbar(cont0, ax = axes) # Add colorbar
lbl0 = axes.set_title("Noon") # set the title for each plot
```

### Bob Wigner: need to trace out Alice

```python
dim = 6
noon = ( tensor( fock(dim,5), fock(dim,0) ) + tensor( fock(dim,0), fock(dim,5) ) ).unit()

# Initalize plot variables
xvec = linspace(-4,4,200)
yvec = linspace(-4,4,200)

wi = wigner(noon.ptrace(1), xvec, yvec, g = 2.0) # Calculate the Wigner function for the states

fig, axes = plt.subplots(1, 1, figsize=(10, 4)) # Create a 1-plot figure
# Contour plots using the results of the Wigner function
cont0 = axes.contourf(xvec, yvec, wi, 200, cmap = 'seismic', norm = plt.Normalize(vmin = -2/pi, vmax = 2/pi))
axes.set_aspect(1) # Make plots squared
fig.colorbar(cont0, ax = axes) # Add colorbar
lbl0 = axes.set_title("Noon") # set the title for each plot
```


### Watch out! the previous Wigner functions do not show any entanglement between Alice and Bob, are just a mixture of $ |0 \rangle \langle 0 |$ and $|5\rangle \langle 5 |$ because we are calculating $ W (\beta) = \frac{2}{\pi} \text{Tr} \left( D^\dagger(\beta) \rho_A D(\beta) \hat{P} \right) $ with $\rho_A = \frac{1}{2}  |0\rangle \langle 0|  + \frac{1}{2}  |5\rangle \langle 5|$


```python
# Plot the Qfunction for a fock vacuum state

# Initialize the states
dim = 30 # Levels in the Hilbert space

n0 = fock(dim, 0)
n5 = fock(dim, 5)

# Initalize plot variables
xvec = linspace(-3,3,200)
yvec = linspace(-3,3,200)

# Calculate the Wigner function for the states
wi_n0 = wigner(n0, xvec, yvec, g = 2.0)
wi_n5 = wigner(n5, xvec, yvec, g = 2.0)
wi_mixture = 1/2 * wi_n0 + 1/2 * wi_n5

# Create a 4-plot figure
fig, axes = plt.subplots(1, 3, figsize=(14, 3))

# Contour plots using the results of the Wigner function
cont0 = axes[0].contourf(xvec, yvec, wi_n0, 100, cmap = 'seismic', norm = plt.Normalize(vmin = -2/pi, vmax = 2/pi))
cont1 = axes[1].contourf(xvec, yvec, wi_n5, 100, cmap = 'seismic', norm = plt.Normalize(vmin = -2/pi, vmax = 2/pi))
cont2 = axes[2].contourf(xvec, yvec, wi_mixture, 100, cmap = 'seismic', norm = plt.Normalize(vmin = -2/pi, vmax = 2/pi))

# Make plots square
axes[0].set_aspect(1)
axes[1].set_aspect(1)
axes[2].set_aspect(1)

# Add colorbar
fig.colorbar(cont0, ax = axes[0])
fig.colorbar(cont1, ax = axes[1])
fig.colorbar(cont2, ax = axes[2])

# Title for each plot
lbl0 = axes[0].set_title("n=0")
lbl1 = axes[1].set_title("n=5")
lbl2 = axes[2].set_title("mixed")
```

# 3. Practice 1: Quantum Teleportation

### Scenario: Alice wants to teleport a state $|\psi \rangle_A = \frac{3}{5} |0 \rangle + \frac{4}{5} |1 \rangle $ to Bob and they share an entangled state  $|\psi \rangle_{AB} = \frac{|00 \rangle_{AB} ~+~ |11 \rangle_{AB}}{\sqrt{2}}  $

### Define states and operators

### Hint: What u need to define
1. bell
2. psi_A
3. cnot 
4. I
5. Z
6. X
7. hadamard
8. proj_00, proj_01, proj_10, proj_11

```python
zero = fock(2,0)
one = fock(2,1)
I = qeye(2)
X = sigmax()
Z = sigmaz()

psi_C = 3/5 * zero + 4/5 * one
ket2dm(psi_C)
```

```python
bell_AB = bell_state('00')
bell_AB
```

```python
cnot = tensor( ket2dm(zero), I) + tensor( ket2dm(one), X)
cnot
```

```python
hadamard = 1/sqrt(2) * Qobj([[1,1],[1,-1]])
hadamard
```

```python
# tensor (state, bell A, bell B) 

initial = tensor( psi_C, bell_AB) #1

psi2 = tensor(cnot, I) * initial #2

psi3 = tensor( hadamard, I, I) * psi2 #3

psi3
```

### Projection and correction

```python
proj_00 = tensor(ket2dm(zero), ket2dm(zero), I)

final = (proj_00 * psi3).ptrace(2).unit()

final
```

```python
proj_01 = tensor(ket2dm(zero), ket2dm(one), I)

final = (proj_01 * psi3).ptrace(2).unit()

X*final*X
```

```python
proj_10 = tensor(ket2dm(one), ket2dm(zero), I)

final = (proj_10 * psi3).ptrace(2).unit()

Z*final*Z
```

```python
proj_11 = tensor(ket2dm(one), ket2dm(one), I)

final = (proj_11 * psi3).ptrace(2).unit()

X*Z*final*Z*X
```

## 4.1 Energy level splitting

Consider an atom (two-level system) with hamiltonian $ H_{qubit} = \left(\begin{array}{cc} 0 & 0 \\ 0 & \hbar \omega_q \end{array}\right) =  \frac{\hbar \omega_q}{2} \left( I - Z \right) $

inside a cavity (resonator) with hamiltonian $ H_{cavity} = \hbar \omega_c \left( a^\dagger a + \frac{1}{2} \right) $ 

whose interaction is given by the hamiltonian $ H_{int} = g \left( a \sigma_+ + a^\dagger \sigma_- \right) $ ,

where $g$ is the strengh of the interaction and $ \sigma_+ = \left( \begin{array}{cc} 0 & 0 \\ 1 & 0 \end{array} \right) $ and
$ \sigma_- = \left( \begin{array}{cc} 0 & 1 \\ 0 & 0 \end{array} \right) $ are the excitation and decay matrices for the qubit, such that $ \sigma_+ |0 \rangle = |1 \rangle $ and $ \sigma_- |1 \rangle = |0 \rangle $.

Energy levels of the full system:



### 4.1.1 Define operators

```python
dim = 4 # dimensionality truncation of cavity

I = qeye(2)
Ic = qeye(dim)

a = destroy(dim)
a_dag = a.dag()

minus = ( sigmax() + 1j * sigmay() )/2 # sigma- operator
plus = minus.dag() # sigma+ operator
```

### 4.1.2 Define Hamiltonian and return energies

Convention: $|\text{cavity} \rangle \otimes | \text{qubit} \rangle $

```python
def energies (Wq,Wc,g): #define parameters that u need to define after 
    
    Hq = Wq/2 * (I - sigmaz()) # Hamiltonian of the qubit, hbar=1
    
    Hc = Wc * ( num(dim) + 1/2 ) # Hamiltonian of the cavity using truncated number operator num(N)
    
    Hint = g * ( tensor (a, plus) + tensor(a_dag, minus) ) # Interaction Hamiltonian
    
    Hfull = tensor(Ic, Hq) + tensor(Hc, I) + Hint # Full Hamiltonian
    
    return Hfull.eigenenergies() # Get eigenvalues
```

### 4.2.1 We consider the resonance case where $ \omega_q = \omega_c $ and without interaction g

```python
# set parameters
g = 0 # No interaction
Wc = 1
Wq = 1

# Figure settings
fig, axes = plt.subplots(1,1)
fig.set_size_inches(7,5)

for i, color in enumerate(['blue','red','red','green','green','purple', 'purple'], start=0):
    plt.scatter(g, energies(Wq, Wc, g) [i], marker = '_', linewidth = 4, color = color )

# Plot labels   
axes.text(.005, .45, r'$ |0g \rangle $',color='b', fontsize=15)
axes.text(.005, 1.45, r'$ |0e \rangle,|1g \rangle  $',color='r', fontsize=15)
axes.text(.005, 2.45, r'$ |1e \rangle,|2g \rangle $',color='g', fontsize=15)
axes.text(.005, 3.45, r'$ |2e \rangle,|3g \rangle $',color='purple', fontsize=15)
      
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("g", fontsize=20)
plt.ylabel("E", fontsize=20)
plt.show()
```

### 4.2.2 Since the qubit and the cavity are in resonance, the energy levels are degenerate. If now we turn on the interaction g

```python
gs=[] #coupling list for each evergy level 
ens=[] #energy list from each coupling
for g in linspace(0, 0.2, 10): # define different g
    gs.append(g * ones(7)) # for every level
    ens.append(energies(Wq, Wc, g))

# print(ones(2*dim))
# print(linspace(0, 0.2, 10))
# print("gs:", gs)
# print("transpose:", transpose(gs))
```

```python
gs=[]
ens=[]

for g in linspace(0, 0.2, 10): # define different g
    gs.append(g * ones(7)) # for every level
    ens.append(energies(Wq, Wc, g))

gs=transpose(gs)    
ens=transpose(ens)    

fig, axes = plt.subplots(1,1)
fig.set_size_inches(10,5)

for i, color in enumerate(['blue','red','red','green','green','purple', 'purple'], start=0):
    plt.plot(gs[i], ens[i], '-',linewidth = 3, color=color)
    
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("Coupling g", fontsize=20)
plt.ylabel("Engergy E", fontsize=20)
plt.show()
```

### In the presence of an interaction, the degerate energy levels split into the eigenstates $ |\phi^n_\pm \rangle = \frac{|g,n+1 \rangle ~ \pm ~ |e,n \rangle}{\sqrt{2}} $ with $E_{\phi_+} \gt E_{\phi_-} $ and the splitting is proportional to $\sqrt{n}$.

### Even if we introduce an excited atom into an empty cavity, the vacuum state of the cavity will split the energy levels of our atom (let that sink in for a moment).

# 5. Unitary Evolution (Ex 2: Spin in a magnetic field)


### Schrodinger equation: $~~ i \hbar \frac{\partial}{\partial{t}} |\psi\rangle = H |\psi\rangle ~~~~$  &rarr;  $ ~~~~\frac{\partial{|\psi\rangle}}{\partial{t}} \frac{1}{|\psi\rangle} = - \frac{i}{\hbar} H ~~~~ $ &rarr; $ ~~~~|\psi(t)\rangle = e^{ - \frac{i H t}{\hbar}} |\psi(0)\rangle = U(t) |\psi(0)\rangle $

### Von Neumann equation: $~~  i \hbar \frac{\partial{\rho}}{\partial{t}}  = [ H,\rho ] $

### Lindblad Master Equation: $~~ i \hbar \frac{\partial{\rho}}{\partial{t}} = [ H,\rho ] + i \hbar \sum_j \gamma_j \left( L_j \rho L_j^\dagger - \frac{1}{2}  \{ L_j^\dagger L_j, \rho \} \right)  $  where $L_j$ are the Lindblad jump operators.

### QuTip deals with all these equation with the single function $ mesolve(H, \rho, t, \{L_j\}_j, \hat{O}) $, where $\hat{O}$ are the operators we want to extract the expectation value of.

## 5.1 Spin in a magnetic field

### We have a spin aligned in the Z-direction in an initial state $ |\psi(0) \rangle = |0 \rangle $ and at time $t=0$ we turn on a magnetic field in the X-direction $ H_{Pauli} = - \vec{\mu} \cdot \vec{B} = \frac{|g| \mu_N B}{2} \sigma_x $. How does the state evolve in time?

### 5.1.1 Define Hamiltonian, states, and operators

```python
H = 1*sigmax() # for clarity let's forget about the constants, which only modify the time scale of the evolution

psi0 = fock(2, 0) # initial state

steps = 100
tlist = linspace(0, 10, steps)

result = mesolve(H, psi0, tlist, [], [])

result.states[0]
```

### 5.1.2 To see the evolution we can plot $\langle \sigma_z \rangle$  as a function of time

```python
fig, axes = plt.subplots(figsize=(10,5))

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
axes.plot( tlist, expect(sigmaz(), result.states) ) #expect(oper, state)
axes.set_xlabel(r'$t$', fontsize=20)
axes.set_ylabel(r'$\left<\sigma_z\right>$', fontsize=20);
```

### 5.1.3 The spin oscillates between $|0 \rangle$ and $ |1 \rangle$. There is a faster way to get $ \langle \sigma_z \rangle$ :

```python
result2 = mesolve(H, psi0, tlist, [], [sigmax(), sigmay(), sigmaz()]) # now we specify the expectation values we want to get

fig, axes = plt.subplots(figsize=(10,5))

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

axes.plot(result.times, result2.expect[2], label=r'$\left<\sigma_z\right>$') # sigma_z
axes.plot(result.times, result2.expect[1], label=r'$\left<\sigma_y\right>$') # sigma_y
axes.plot(result.times, result2.expect[0], label=r'$\left<\sigma_x\right>$') # sigma_x

axes.set_xlabel(r'$t$', fontsize=20)
axes.legend(fontsize=15);
```

# 6. Time-dependent hamiltonian (Ex 3: Evolution of a cat state)

Plot the wigner function of a cat state after a evolution time  

## 6.1 Define states, operators, Hamiltonian, and plot wigner

```python
dim = 18
alpha = 2
select_time = 31

a = destroy(dim)
chi = 1
H = chi/2 * (a.dag() * a + 1/2)
time = linspace(0,pi,32)
cat = ( coherent(dim, alpha) + coherent(dim, -alpha) ).unit()

# Initalize plot variables
xvec = linspace(-3,3,100)
yvec = linspace(-3,3,100)

result = mesolve(H, cat, time, [], [])


# plot wigner 
wig = wigner(Qobj(result.states[select_time]), xvec, yvec, g = 2.0)


fig, axes = plt.subplots(1, 1, figsize=(10, 4)) # Create a 1-plot figure
cont0 = axes.contourf(xvec, yvec, wig, 200, cmap = 'seismic', norm = plt.Normalize(vmin = -2/pi, vmax = 2/pi))
axes.set_aspect(1) # Make plots squared
fig.colorbar(cont0, ax = axes) # Add colorbar
lbl0 = axes.set_title("Cat") # set the title for each plot
# print(time)
# print(select_time)
```

## 6.2 Want to see how it evolve?

```python
dim = 18

alpha = 2

a = destroy(dim)

chi = 2

H = chi/2 * (a.dag() * a + 1/2)

cat = ( coherent(dim, alpha) + coherent(dim, -alpha) ).unit()

tlist = linspace(0,pi,32)

result = mesolve(H, cat, tlist, [], [])
```

```python
xvec = linspace(-4,4,200)
yvec = linspace(-4,4,200)

# Array of wigner functions
wigner_list = [wigner(state, xvec, yvec, g = 2.0) for state in result.states] 

# Figure settings
fig, axes = plt.subplots(1,1)
fig.set_size_inches(10,8)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel(r'Re ($\alpha$)', fontsize=20)
plt.ylabel(r'Im ($\alpha$)', fontsize=20)
plt.title("Cat state", fontsize=20)

# Color bar settings
wigner0 = wigner(result.states[0], xvec, yvec)
cont0 = axes.contourf(xvec, yvec, wigner0, 200, cmap = 'seismic', norm = plt.Normalize(vmin = -2/pi, vmax = 2/pi))
fig.colorbar(cont0)

def wigner_frame(frame): # gets a frame number and ouputs a wigner plot
    wi = wigner_list[frame]
    cont = axes.contourf(xvec, yvec, wi, 200, cmap = 'seismic', norm = plt.Normalize(vmin = -2/pi, vmax = 2/pi))

anim = FuncAnimation(fig, wigner_frame, frames = len(wigner_list), interval = 150) # gif maker

anim.save('EvolvingCat.gif', writer='imagemagick')
```

# 7. Joint Wigner Tomography

### Joint Wigner Tomography: In order to visualize the effects of the entanglement we need to perform Wigner of two modes. 
### However, since single-mode Wigner have 2 dimensions ($\text{Re}(\beta), \text{Im}(\beta)$), with 2 modes we have 4 dimensions <br />   ($\text{Re}(\beta_A), \text{Im}(\beta_A), \text{Re}(\beta_B), \text{Im}(\beta_B) $). 
### We cannot plot all of them in a 2D plot, so we have to choose which dimensions, typically one plot of <br />($\text{Re}(\beta_A), \text{Re}(\beta_B)$) and another of ($\text{Im}(\beta_A), \text{Im}(\beta_B)$).

### Finally, 2-mode Wigner function is not implemented in QuTip, so we have to define it manually as: 
### $W_J (\beta_A, \beta_B) = \frac{4}{\pi} \text{Tr} \left(  D^\dagger(\beta_B) D^\dagger(\beta_A) \rho D(\beta_A) D(\beta_B) \hat{P_J}\right)$ with the joint parity $ \hat{P_J} = e^{~i \pi (\hat{n}_A + \hat{n}_B)} = e^{~i \pi ~(\hat{n}_A \otimes I_B + I_A \otimes \hat{n}_B)} $



```python
dim = 24

I = qeye(dim)

noon = ( tensor( fock(dim,5), fock(dim,0) ) + tensor( fock(dim,0), fock(dim,5) ) ).unit() # NOON state

P = ( 1j * np.pi * (tensor(I, num(dim))+ tensor(num(dim),I)) ).expm() # joint parity

# Initalize plot variables
points = 20

list_beta_A = np.linspace(-3,3,points)

list_beta_B = np.linspace(-3,3,points)

# List of results
results = np.zeros(points**2)


for idx, (beta_A, beta_B) in enumerate(list(itertools.product(list_beta_A, list_beta_B))):
    
        D = tensor( displace(dim, beta_A), displace(dim, beta_B) ) # To get the ( Re(\beta_A), Re(\beta_B) ) plot
        #D = tensor( displace(dim, 1j*beta_A), displace(dim, 1j*beta_B) ) # To get the ( Im(\beta_A), Im(\beta_B) ) plot
        
        results[idx] = (2/np.pi)**2 * ( ket2dm( D.dag() * noon ) * P ).tr()

# Plot results        
fig, axes = plt.subplots(1, 1, figsize = (10, 4))

cont0 = axes.contourf(list_beta_A, list_beta_B, results.reshape((points, points)), 200, cmap = 'seismic', norm = plt.Normalize(vmin = -(2/np.pi)**2, vmax = (2/np.pi)**2))

axes.set_aspect(1) 

fig.colorbar(cont0, ax = axes) 

lbl0 = axes.set_title("Joint Wigner")   
```

# 9. Time-dependent hamiltonian: There is a weak magnetic field rotating in time

```python


wo = 3 # frequency (energy) of our spin

H0 = wo/2*sigmaz() # constant vertical magnetic field

H1 = sigmax()  # oscillating field in x direction

H2 = sigmay() # oscillating field in y direction

g = 0.1 # strengh of the oscillating field

psi0 = fock(2, 0) # initial state

t = linspace(0, 40, 100) # list of times for the time evolution
     
max_prob=[]    
    
wlist = linspace (2,4,100) # we sweep the frequency
    
for w in wlist:

    def H1_coeff(t, args):
        return g/2 * cos(w*t)

    def H2_coeff(t, args):
        return g/2 * sin(w*t)   
        
    H = [ H0, [H1, H1_coeff], [H2, H2_coeff] ] # Hamiltonians that depend on time: [constant part, coefficient]
       
    result = mesolve(H, psi0, t, [], [ket2dm(fock(2,1))]) # probability of excitation as a function of time
    
    max_prob.append(max(result.expect[0])) # maximum probability of excitation for a given frequency

fig, axes = plt.subplots(figsize=(10,5))
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
axes.plot( wlist, max_prob)
axes.set_xlabel(r'$\omega$', fontsize=20)
axes.set_ylabel(r'$P_1$' , fontsize=20);

```

### When the oscillating magnetic field frequency $\omega$ matches the frequency of the given atom $\omega_0$, we have a resonance where the atoms flip their state. We can then detect the changes in magnetization using a coil. This is nothing else than a nuclear magnetic resonance (NRM)!
