```python
def state_different_phase(t_wait_phase, ALPHA=1): 
    t_displace = np.linspace(0, 0.03, 2)
    t_wait = np.linspace(0, 0.1, 2)
    EPSILON = 64
    d_alpha = 2
#     t_wait_phase = np.linspace(0, 0.01, 2)
    state_dif_phase = []
    state_1 = ECD_ideal_operator(superpos, 1j*ALPHA) #ECD_exp(superpos, t_displace, t_wait, ALPHA, epsilon = EPSILON)
    state_2 = D_ideal_operator(-ALPHA/2*1j)*state_1 # one goes to vac, one goes to -2alpha D_ideal(superpos, t_displace, epsilon = d_alpha/2*EPSILON) #D(superpos,t_displace, epsilon = d_alpha/2*EPSILON)
    state_3 = phase_ideal_operator(state_2, t_wait_phase) #qt.mesolve(H_dispersive(), state_2, t_wait_phase).states[-1] # wait for accumating phase
    state_4 = D_ideal_operator(+ALPHA/2*1j)*state_3 #displace back
    state_5 = ECD_ideal_operator(state_4, 1j*ALPHA) # ecd back too vac ECD_exp(state_4, t_displace, t_wait, +ALPHA, epsilon = EPSILON) ##question
    state_6 = tensor(hadamard,Ic)*state_5
    return state_6
```
