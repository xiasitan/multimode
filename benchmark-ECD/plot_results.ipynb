```python
import os
import json
import pylab

sourcedir = "C:/Users/jonat/Desktop/Code/multimode/benchmark-ECD/results"


def transform_version(points):
    # turn a list of dicts into two lists of values
    Nvec = []
    tvec = []
    for x in points:
        Nvec.append(x["N"])
        tvec.append(x["t"])
    return Nvec, tvec


def transform_json(d):
    data = []
    for name, points in d.items():
        data.append((name, transform_version(points)))
    #data.sort(key=lambda x: x[0])
    return data
```

```python
filename = 'julia-ecd_timeevolution_master_1.json'
f = open(os.path.join(sourcedir, filename))
d = json.load(f)
f.close()

def transform_json_to_data(d):
    Nvec = []
    tvec = []
    for data_d in d:
        Nvec.append(data_d["N"])
        tvec.append(data_d["t"])
    return Nvec, tvec

Nvec, tvec = transform_json_to_data(d)
pylab.title(filename)
pylab.plot(Nvec,tvec, ".")
pylab.legend()
pylab.xlabel("Cutoff Dimension")
pylab.ylabel("Runtime in (s)")
        

# pylab.figure()
# for name, points in data:
#     pylab.title(testname)
#     pylab.plot(points[0], points[1], label=name)
#     pylab.plot(points[0], points[1], "ok", alpha=0.4)
# pylab.legend()
# pylab.show()
```
